```python
import pandas as pd
import seaborn as sns
```

```python
data=pd.read_csv('mushrooms.csv')
```

```python
data.describe()
```

```python
#converting categorical values to numerical values
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for column in data.columns:
    data[column]=le.fit_transform(data[column])

```

```python
data.head()
```

```python
#response target as y , feature matrix in x 
X=data.drop("class",axis=1)
y=data['class']
print(y)

```

```python
# Using heatmap to find corelation  
import matplotlib.pyplot as plt
plt.figure(figsize=(15, 15))
sns.heatmap(data.corr(),annot=True,cmap='viridis',linewidths=0.5)
```

```python
#Applying PCA (principal component analysis)
from sklearn.decomposition import PCA
pca1=PCA(n_components=7)
pca_fit=pca1.fit_transform(X)
```

```python
pca1.explained_variance_ratio_
sum(pca1.explained_variance_ratio_)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(pca_fit,y,test_size=0.20,random_state =41)
```

```python
y_train
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

```python
lr=LogisticRegression()
lr.fit(X_train,y_train)

knn=KNeighborsClassifier()
knn.fit(X_train,y_train)

svc=SVC()
svc.fit(X_train,y_train)

dt=DecisionTreeClassifier()
dt.fit(X_train,y_train)

rm=RandomForestClassifier()
rm.fit(X_train,y_train)

gb=GradientBoostingClassifier()
gb.fit(X_train,y_train)
```

```python
y_pred1=lr.predict(X_test)
y_pred2=knn.predict(X_test)
y_pred3=svc.predict(X_test)
y_pred4=dt.predict(X_test)
y_pred5=rm.predict(X_test)
y_pred6=gb.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print('ACC LR',accuracy_score(y_test,y_pred1))
print('ACC KNN',accuracy_score(y_test,y_pred2))
print('ACC SVC',accuracy_score(y_test,y_pred3))
print('ACC DT',accuracy_score(y_test,y_pred4))
print('ACC RM',accuracy_score(y_test,y_pred5))
print('ACC GBC',accuracy_score(y_test,y_pred6))
#here it can be concluded that Random forest classifier gives the most accurate result
```
